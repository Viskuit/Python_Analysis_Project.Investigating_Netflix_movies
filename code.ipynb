```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

netflix_df = pd.read_csv("./netflix_data.csv") # First we load the data
netflix_subset = netflix_df[netflix_df["type"] != "TV Show"] # We filter out "TV Show" occurences
netflix_movies = netflix_subset[["title", "country", "genre", "release_year", "duration"]] # We select title, country, genre, release_year, and duration columns

short_movies = netflix_movies[netflix_movies["duration"] < 60] # Movies with a duration less than 60 minutes

# In the next part we divide the genres by colors    
genre_color_mapping = {
    "Children": "Blue",
    "Documentaries": "Green", 
    "Stand-Up": "Purple"
}
colors = [genre_color_mapping.get(row["genre"], "Grey") for _, row in netflix_movies.iterrows()] # list comprehension. "Grey" for everything else.

# Now let's make the plot
fig = plt.figure()
plt.scatter(data=netflix_movies, x="release_year", y="duration", c=colors)
plt.xlabel("Release year")
plt.ylabel("Duration (min)")
plt.title("Movie Duration by Year of Release")
plt.show()

# Are we certain that movies are getting shorter?"
answer = "no"
```
